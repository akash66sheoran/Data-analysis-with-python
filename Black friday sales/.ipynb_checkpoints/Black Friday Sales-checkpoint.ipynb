# Analyzing Gender

```python
data.groupby('Gender')['Purchase'].mean()
```

```python
data['Gender'].value_counts()
```

```python
genderData = pd.DataFrame({'Ratio' : [len(data[data['Gender'] == 'M']), len(data[data['Gender'] == 'F'])]}, index = ['Male', 'Female'])
```

```python
genderData.plot.pie(y = 'Ratio', figsize = (6,6), autopct = "%.1f")
```

```python
data['Gender'].value_counts().plot(kind = 'pie', figsize = (6,6), autopct = "%.1f", title = 'Gender Ratio')
```

```python
data.groupby('Gender')['Purchase'].mean()
```

```python
data.groupby('Gender')['Purchase'].mean().plot(kind = 'bar', figsize = (6,6), title = 'Average money spent by a male/female')
```

# Analyzing Age

```python
data.groupby('Age').size().plot(kind = 'bar', figsize = (12,6), title = 'Purchase distribution by age')
```

```python
data.groupby('Age')['Purchase'].mean().plot.bar(figsize = (12,6))
```

```python
data.groupby('Age')['Purchase'].sum().plot.bar(figsize = (12,6))
```

```python
data.groupby('Age')['Product_ID'].nunique().plot.bar(figsize = (12,6))
```

```python
sns.set(rc = {'figure.figsize': (12,6)})
sns.countplot(x = 'Age', hue = 'Gender', data = data)
```
